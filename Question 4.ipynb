```python
import math
import matplotlib.pyplot as plt

def time_to_fill(initial_volume, target_volume, doubling_time=5):
    """
    Calculate the time required to fill a target volume when an object doubles in size every `doubling_time` minutes.

    Args:
    - initial_volume (float): Initial volume of the object in m^3.
    - target_volume (float): Volume of the space to fill in m^3.
    - doubling_time (float): Time taken for the volume to double, in minutes.

    Returns:
    - time (int): Total time required to fill the space in minutes.
    - volumes (list): List of volumes at each doubling step.
    - times (list): List of times at each doubling step.
    """
    current_volume = initial_volume
    time = 0
    volumes = [current_volume]
    times = [time]
    
    while current_volume < target_volume:
        time += doubling_time
        current_volume *= 2
        volumes.append(current_volume)
        times.append(time)
    
    return time, volumes, times

# Constants
volume_chestnut_bun = 2e-4  # Initial volume of chestnut bun in m^3
radius_solar_system = 1.496e13  # Radius of Solar System in meters
volume_solar_system = (4 / 3) * math.pi * radius_solar_system**3  # Volume in m^3
volume_tokyo_dome = 1.24e6  # Volume of Tokyo Dome in m^3

# Time to fill the Solar System
time_solar_system, volumes_solar_system, times_solar_system = time_to_fill(volume_chestnut_bun, volume_solar_system)

# Time to fill Tokyo Dome
time_tokyo_dome, volumes_tokyo_dome, times_tokyo_dome = time_to_fill(volume_chestnut_bun, volume_tokyo_dome)

# Results
print(f"Time to fill the Solar System: {time_solar_system} minutes.")
print(f"Time to fill the Tokyo Dome: {time_tokyo_dome} minutes.")

# Plotting growth over time
plt.figure(figsize=(12, 6))

# Plot for Solar System
plt.subplot(1, 2, 1)
plt.plot(times_solar_system, volumes_solar_system, label="Solar System", color="orange")
plt.xlabel("Time (minutes)")
plt.ylabel("Volume (m^3)")
plt.title("Growth of Chestnut Bun Volume (Solar System)")
plt.yscale("log")
plt.legend()

# Plot for Tokyo Dome
plt.subplot(1, 2, 2)
plt.plot(times_tokyo_dome, volumes_tokyo_dome, label="Tokyo Dome", color="green")
plt.xlabel("Time (minutes)")
plt.ylabel("Volume (m^3)")
plt.title("Growth of Chestnut Bun Volume (Tokyo Dome)")
plt.yscale("log")
plt.legend()

plt.tight_layout()
plt.show()
```
